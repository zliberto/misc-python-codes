### Exercise 7

```python
# set the result column to the value of "Tie", then change each row of Result depending on the score
df2016["Result"] = "Tie"
df2016.loc[df2016["UIC Score"] > df2016["Opp Score"],"Result"] = 'Win'
df2016.loc[df2016["UIC Score"] < df2016["Opp Score"], "Result"] = 'Loss'

# print(df2016[["Opponent", "Result"]])

# group by result and count

result_group = df2016.groupby("Result")

result_count = result_group["Result"].count()

# this line of code is here so that in case there is no ties (like in this case), it displays "Tie = 0"
result_count.loc["Tie"] = df2016["Opponent"].count() - result_count.loc["Win"] - result_count.loc["Loss"]


print("\n\n",result_count)
```
